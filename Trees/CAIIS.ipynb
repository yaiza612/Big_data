```python
test = spark.read.options(header='True', inferSchema='True').csv('/content/drive/MyDrive/Colab_Notebooks/test.csv')
train = spark.read.options(header='True', inferSchema='True').csv('/content/drive/MyDrive/Colab_Notebooks/train.csv')
```

```python
from pyspark.ml import Pipeline
from pyspark.ml.feature import StringIndexer
categoricalColumns = [item[0] for item in train.dtypes if item[1].startswith('string') ]

stages = []


for categoricalCol in categoricalColumns:
    
    stringIndexer = StringIndexer(inputCol = categoricalCol, outputCol = categoricalCol + 'Index')

    
    stages += [stringIndexer]


pipeline = Pipeline(stages = stages)

pipelineModel = pipeline.fit(train)

train= pipelineModel.transform(train)
```

```python
categoricalColumns = [item[0] for item in test.dtypes if item[1].startswith('string') ]

stages = []


for categoricalCol in categoricalColumns:
    
    stringIndexer = StringIndexer(inputCol = categoricalCol, outputCol = categoricalCol + 'Index')

    
    stages += [stringIndexer]


pipeline = Pipeline(stages = stages)

pipelineModel = pipeline.fit(test)

test= pipelineModel.transform(test)
```

```python
from pyspark.ml.feature import VectorAssembler, Imputer
assembler = VectorAssembler(
    inputCols=[
 'cont_0',
 'cont_1',
 'cont_2',
 'cont_3',
 'cont_4',
 'cont_5',
 'cont_6',
 'cont_7',
 'cont_8',
 'cont_9',
 'cont_10',
 'cat_0Index',
 'cat_1Index',
 'cat_2Index',
 'cat_3Index',
 'cat_4Index',
 'cat_5Index',
 'cat_6Index',
 'cat_7Index',
 'cat_8Index',
 'cat_9Index',
 'cat_10Index',
 'cat_11Index',
 'cat_12Index',
 'cat_13Index',
 'cat_14Index',
 'cat_15Index',
 'cat_16Index',
 'cat_17Index',
 'cat_18Index'],
 outputCol='features')
output = assembler.transform(train)
imputer = Imputer(inputCols=['target'], outputCols=['label'], strategy='mean')
imputer_model = imputer.fit(output)
final_train = imputer_model.transform(output)
```

```python
final_train.show()
```

```python
from pyspark.ml.classification import (RandomForestClassifier, GBTClassifier, DecisionTreeClassifier)
dtc = DecisionTreeClassifier(maxBins=300) 
rfc = RandomForestClassifier(numTrees = 100, maxBins=300)
gbt = GBTClassifier(maxBins=300)
```

```python
dtc_model = dtc.fit(final_train)
rfc_model = rfc.fit(final_train)
gbt_model = gbt.fit(final_train)
```

```python
from pyspark.ml.feature import VectorAssembler
assembler = VectorAssembler(
    inputCols=[
'cont_0',
'cont_1',
'cont_2',
'cont_3',
'cont_4',
'cont_5',
'cont_6',
'cont_7',
'cont_8',
'cont_9',
'cont_10',
'cat_0Index',
 'cat_1Index',
 'cat_2Index',
 'cat_3Index',
 'cat_4Index',
 'cat_5Index',
 'cat_6Index',
 'cat_7Index',
 'cat_8Index',
 'cat_9Index',
 'cat_10Index',
 'cat_11Index',
 'cat_12Index',
 'cat_13Index',
 'cat_14Index',
 'cat_15Index',
 'cat_16Index',
 'cat_17Index',
 'cat_18Index'],
     outputCol='features')
final_test = assembler.transform(test)
```

```python
dtc_preds = dtc_model.transform(final_test)
rfc_preds = rfc_model.transform(final_test)
gbt_preds = gbt_model.transform(final_test)
```

To evaluate the models we need to split train

```python
train_x, train_y = final_train.randomSplit([0.9, 0.1])
```

```python
dtc_model = dtc.fit(train_x)
rfc_model = rfc.fit(train_x)
gbt_model = gbt.fit(train_x)
```

```python
dtc_preds_train = dtc_model.transform(train_y)
rfc_preds_train = rfc_model.transform(train_y)
gbt_preds_train = gbt_model.transform(train_y)
```

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
evaluator = MulticlassClassificationEvaluator(metricName='accuracy')
```

```python
print(f'DTC: {evaluator.evaluate(dtc_preds_train)} \n Features Importance: {dtc_model.featureImportances}\n\n') 
print(f'RFC: {evaluator.evaluate(rfc_preds_train)} \n Features Importance: {rfc_model.featureImportances}\n\n')
print(f'GBT: {evaluator.evaluate(gbt_preds_train)} \n Features Importance: {gbt_model.featureImportances}\n\n')
```
